```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('user_data.csv')

# Convert the income column to numeric, setting errors='coerce' to handle any non-numeric values
df['income'] = pd.to_numeric(df['income'], errors='coerce')

# Group by age and sum the income, then plot as a bar chart
age_income = df.groupby('age')['income'].sum()
age_income.plot(kind='bar')
plt.title('Ages with Highest Income')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.savefig('age_income.png', format='png', dpi=300) # Save the plots as PNG images for use in a PowerPoint presentation
plt.show()


# Melt the DataFrame for easier plotting
df_melted = df.melt(id_vars=['gender'], value_vars=['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare'], 
                    var_name='category', value_name='amount')

# Create a bar plot to show spending by gender and category
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='amount', hue='gender', data=df_melted)
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Spending Category')
plt.ylabel('Amount')
plt.savefig('gender_spending.png', format='png', dpi=300) # Saves the plots as PNG images for use in a PowerPoint presentation
plt.show()

```
